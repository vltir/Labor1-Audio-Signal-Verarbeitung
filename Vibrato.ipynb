# Vibrato Effekt

[Vibrato](https://de.wikipedia.org/wiki/Vibrato) bezeichnet eine periodische Schwankung der Tonhöhe (Frequenzmodulation). Die Frequenz dieser periodischen Schwankung liegt gewöhnlich unterhalb 20 Hz.

Ähnlich wie beim Doppler-Effekt wird die zu verarbeitende Audiodatei verzögert eingelesen. Allerdings ist die Verzögerung nicht stets konstant und soll fortschreitend variieren. Die Länge der **momentanen Verzögerung** gibt die Amplitude eines LFO vor.

## Mehr zum Thema

Insbesondere beim Gesang wird unter Vibrato nicht nur ausschließlich eine Frequenzschwankung, sondern auch Amplitudenschwankung oder sogar beides verstanden. Insbesondere bei Blasinstrumenten handelt es sich um Frequenz-Vibrato, welches je nach Instrument auf verschiedene Art und Weise ausgeübt werden kann.

Vibrato dient einerseits dazu, den Ton "lebendiger" zu gestalten oder auch die Intonationsungereimtheiten zu "kaschieren".
