## Hintergrund

Die einzelnen Tonspuren eines Songs liegen teils als Monospur vor. Beispielsweise durch ein Mikrofon abgenommenen Gesang oder diverse Instrumente. Bei manchen Instrumenten wie z.B. Klavier und Schlagzeug wird oft aus akustischen Gründen komplexe mehrspurige Mikrofonierung verwendet. Synthetische Instrumente generieren in der Regel bereits eine Stereospur.

Schlussendlich erstellt ein Toningenieur den finalen Summenmix aus allen einzelnen Tonspuren, welche ggf. mit weiteren Audioeffekten verziert werden. Reine Monospuren werden dabei gewöhnlich dupliziert. Durch die Subtraktion der Stereospuren können identische Signalamplitugen gelöscht werden, wie es bei duplizierten Monospuren der Fall ist.

## Alternativen

Diese, recht einfache aber dennoch willkürliche Methode zur "entvokalisierung" der Stereoaufnahmen ist ferner verwandt mit sog. [Source Separation](https://en.wikipedia.org/wiki/Signal_separation) Techniken.

Teilweise können durch spektrale Signalverarbeitung harmonische und perkussiven Signalkomponenten voneinander isoliert werden (mehr Infos dazu folgen im späteren Vorlesungskapitel).

Außerdem gibt es zahlreiche CNN basierte Ansätze, wie z.B. [Demucs](https://github.com/facebookresearch/demucs) von Facebook Research. Probier's also gerne [online](https://huggingface.co/spaces/akhaliq/demucs) oder [offline](https://github.com/facebookresearch/demucs#separating-tracks) in Python aus.
