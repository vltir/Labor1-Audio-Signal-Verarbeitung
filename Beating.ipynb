# Mono-/Binaural Beats

Generiere zwei unterschiedlich schnelle Sinuswellen, deren Frequenzen nahe beieinander liegen. Je nach Frequenzdifferenz soll eine wahrnehmbare Schwebung entstehen, andernfalls zwei differenzierbare Töne oder Unisono.

Probiere folgende Umsetzungsvarianten aus:

- Summe der beiden Sinuswellen als Monospur
- Beide Sinuswellen getrennt als Stereospur
- Alternatives Rezept aus dem Synthese-Kapitel (siehe Folie 61)

```python
import IPython as ipy
import numpy as np

sr = 44100 # sample rate in hertz
f1 = 440   # signal frequency in hertz
f2 = 444
d = 10     # signal duration in seconds
n = d * sr # signal duration in samples

t = np.arange(0, d, 1/sr) # timestamps in seconds

x1 = np.sin(2 * np.pi * f1 * t) # wave generation
x2 = np.sin(2 * np.pi * f2 * t)
x  = x1 + x2

ipy.display.Audio(x, rate=sr)
```

```python
ipy.display.Audio([x1, x2], rate=sr)
```
