# Chameleon

In diesem Arbeitsblatt geht es darum, eine Melodie zu synthetisieren. Dabei handelt es sich um Basslinie aus dem Musikstück [Chameleon](https://www.youtube.com/watch?v=WYRrIBqKsJ4) von [Herbie Hancock](https://de.wikipedia.org/wiki/Herbie_Hancock) aus dem Jahr 1973.

In der Originalaufnahme wurde diese Melodie auf einem [ARP-Odyssey](https://en.wikipedia.org/wiki/ARP_Odyssey)-Synthesizer gespielt.

**Alternativlösung:** Wenn du möchtest, probiere gerne den Onlinedienst [NoiseCraft](https://noisecraft.app) aus. Lade hierfür die angehängte Datei `Chameleon.ncft` hoch.

```python
from dhbw import dasp
from IPython.display import Audio

import numpy as np
```

Die Maßeinheit *Beats Per Minute* legt das Tempo fest. In der Originalaufnahme ändert sich das Tempo, was hier jedoch nicht berücksichtigt werden muss.

```python
bpm = (1/4) * 100  # quarter beats per minute
```

## Partitur

Die Python-Partitur legt einzelne Töne der Melodie fest:

 - Tonname
 - Oktave
 - Relative Tondauer

Da die Melodie sich permanent wiederholt, genügt bereits ein Teil davon in der Partitur zu beschreiben.

```python
o = 2 # base octave offset

notes = \
[
    # [NOTE, OCTAVE, DURATION]

    # PART A
    ['G',  (o+0), 1/8],
    ['G#', (o+0), 1/8],
    ['A',  (o+0), 1/8],
    ['A#', (o+0), 3/16],
    ['G#', (o+1), 3/16],
    ['A#', (o+1), 1/4],

    # PART B
    ['C',  (o+1), 1/8],
    ['C#', (o+1), 1/8],
    ['D',  (o+1), 1/8],
    ['D#', (o+1), 3/16],
    ['A#', (o+1), 3/16],
    ['C#', (o+2), 1/4],
]
```

# Synthese

In diesem Abschnitt wird die Melodie Ton für Ton (also `sample` für `sample`) generiert.

Noch klingt sie aber etwas langweilig...

Probiere verschiedene Oszillatoren z.B. aus vorherigen Übungen oder aus dem Namespace `dasp.signal`. Probiere auch Modulationstechniken zwecks Klangformung aus.

Gewichte außerdem die Samples mit einer Amplituden-Hüllkurve, z.B. [ADSR](https://de.wikipedia.org/wiki/ADSR), sodass die einzelnen Töne abklingen und nicht mehr "aneinanderkleben".

**Künstlerischer Tipp:**

Bei der FM-Synthese, füttere den Modulator mit der oktavierten Signalfrequenz des Trägers. Das könnte spannend werden! 😉

```python
melody = []

for note in notes:

    semitone, octave = dasp.pitch.tone(note[0]), note[1]
    frequency = dasp.pitch.frequency(semitone, octave)
    duration = note[-1] * (60 / bpm) # seconds
    timeline = dasp.timeline(duration)

    sample = dasp.signal.harmonic(frequency, timeline)

    melody = np.concatenate((melody, sample))
```

```python
loop = np.tile(melody, 3)
```

```python
Audio(loop, rate=dasp.SR)
```
