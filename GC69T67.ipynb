# Hier spielt die Musik!

Ein Kommilitone bittet um deine Hilfe! Er sucht nämlich ein Geocache. Die Koordinate des Verstecks ist in einer Sprachnachricht enthalten. Leider kann die Sprachnachricht nicht direkt abgehört werden...

Die abzuhörende Tonspur ist in einem Bild kodiert. Das Bild visualisiert eine Schallplatte. Die Intensität der [spiralförmig](https://en.wikipedia.org/wiki/Archimedean_spiral) verlaufenden Rille entspricht der Amplitude des Audiosignals.

Somit muss die Tonspurspirale Schritt für Schritt aus dem Bild ausgelesen, um als Audiosignal interpretiert zu werden. Dabei sind die ausgelesenen `8-bit` Bildintensitaten in die Audiosignalamplitude `[-1.0 ... +1.0]` durch konstante Verschiebung und Skalierung passend umzuwandeln.

```python
import IPython as ipy
import matplotlib.pyplot as plot
import numpy as np
import skimage

from dhbw import dasp
```
